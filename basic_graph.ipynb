<h1>Chpter02크리스마스의 기온변화 그리기</h1>

```python
import csv
#window = cp949
#linux&macOS = utf-8
f = open("seoul.csv", 'r', encoding="cp949")
data = csv.reader(f)

header = next(data)
print(header)
rslt = []
mean = [] # 평균 기온 데이터를 저장할 리스트 생성
low = [] # 최저 기온 데이터를 저장할 리스트 생성
high = [] # 최고 기온 데이터를 저장할 리스트 생성
for row in data :
    if '' != row[4]:
        if row[0].split('-')[1] == '12' and row[0].split('-')[2] == '25':
            mean.append(float(row[2]))
            low.append(float(row[3]))
            high.append(float(row[4]))

f.close()
print(len(low))
print(mean)
```

```python
plt.figure(figsize=(12,2))
plt.rc("font", family='Malgun Gothic')
plt.rcParams["axes.unicode_minus"] = False
plt.title('크리스마스의 기온 변화 그래프')
plt.plot(high, color="r", label="max")
plt.plot(mean, color="g", label="mean")
plt.plot(low, color="b", label="min")
plt.legend()
plt.xlabel("일")
plt.ylabel("기온")
plt.grid()
plt.show()
```

<h1>지구온난화 시각화 해보기</h1>

```python
##지구온난화 근거 시각화

f = open("seoul.csv", 'r', encoding="cp949")
data = csv.reader(f)

header = next(data)

year_list = [] # 평균 기온 데이터를 저장할 리스트 생성
day = [] # 365일 기온값을 저장
mean = [] # sum(day) / len(day)
for row in data :
    year = int(row[0].split('-')[0])
    if year not in [1907, 1950, 1951, 1952, 1953, 2024] :
        if '' != row[2]:
            day.append(float(row[2]))
            if (row[0].split('-')[1] == "12") and (row[0].split("-")[2] == "31"):
                mean.append(sum(day) /len(day)) #해당연도의 평균값 저장
                year_list.append(year)
                day = []
            
    elif year == '2024':
        break


f.close()
print(len(low))
```

```python
plt.figure()
plt.plot(year_list, mean, color='r', marker="^", lw=0.5)
plt.grid()
plt.show()
```
